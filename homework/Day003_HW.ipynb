```python
# 下載檔案
import urllib.request
import zipfile

res = "http://opendata.cwb.gov.tw/govdownload?dataid=F-D0047-093&authorizationkey=rdec-key-123-45678-011121314"
urllib.request.urlretrieve(res, "./data/example.zip")
f = zipfile.ZipFile('./data/example.zip')
f.extractall('./data')
```

### 比較一下範例檔案中的「File I/O」與「xmltodict」讀出來的內容有什麼差異


```python
import os, sys

# 打开文件
dirs = os.listdir( './data' )

# 输出所有文件和文件夹
for file in dirs:
    print(file)
```

```python
# 讀檔案
fh = open("./data/64_72hr_CH.xml", "r", encoding='utf-8')
xml = fh.read()
fh.close()

print(xml)


```

### 根據範例檔案的結果：

1. 請問高雄市有多少地區有溫度資料？
2. 請取出每一個地區所記錄的第一個時間點跟溫度
3. 請取出第一個地區所記錄的每一個時間點跟溫度

```python
# 1. 請問高雄市有多少地區有溫度資料？

import xmltodict
d = dict(xmltodict.parse(xml))
# 取出 datasetDescription
locations = d['cwbopendata']['dataset']['locations']['location']

count=0
for loc in locations:
    print(loc['locationName'])
    count+=1
    print(count)

print("\n高雄市有多少地區有溫度資料?:",count)    
    
```

```python
# 2. 請取出每一個地區所記錄的第一個時間點跟溫度

import pandas as pd

import xmltodict
d = dict(xmltodict.parse(xml))
# 取出 datasetDescription
locations = d['cwbopendata']['dataset']['locations']['location']
data = []

for loc in locations:
    
    location = loc['locationName']
    dataTime = loc['weatherElement'][0]['time'][0]['dataTime']
    value = loc['weatherElement'][0]['time'][0]['elementValue']['value']
    data.append({'地區':location,'時間點':dataTime,'攝氏度':value})
 
df=pd.DataFrame(data)
df
```

```python
# 3. 請取出第一個地區所記錄的每一個時間點跟溫度
import xmltodict
d = dict(xmltodict.parse(xml))
# 取出 datasetDescription
locations = d['cwbopendata']['dataset']['locations']['location'][0]
location = locations['locationName']
print(location)

data2 =[]

for loc in locations['weatherElement'][0]['time']:
    
  
    dataTime = loc['dataTime']
    value = loc['elementValue']['value']
    data2.append({'鹽埕區:時間點':dataTime,'攝氏度':value})
 
df2=pd.DataFrame(data2)
df2
```
