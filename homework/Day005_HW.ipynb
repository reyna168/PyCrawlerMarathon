## 作業目標

* 請利用 API: https://www.dcard.tw/_api/forums/pet/posts?popular=true 回答下列問題：

1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

```python
# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？

import requests
import json

r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
data=json.loads(r.text)
count=0
for i in data:
    print(i['id'])
    count+=1
print('總共回傳%s資料'%count)

for key in data[0].keys():
    print(key)
        
```

```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」

count = 1
for d in data:
    print(count,'標題: ',d['title'])
    count+=1
    print('貼文時間: ',d['createdAt'])
    print('留言人數: ',d['commentCount'])
    print('按讚人數: ',d['likeCount'])
    print('================')
```

```python

import pandas as pd

df1 = pd.DataFrame(data)

#選取我們需要什麼東西
popular= df1[['title','createdAt','commentCount','likeCount']]
#將我們需要的東西改個名字
popular.columns = ['標題:','貼文時間:','留言人數:','按讚人數:']
#呈現出來
popular
```

```python
# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」
# 個人備註
#https://www.dcard.tw/_api/forums/pet/posts?popular=true  這是熱門
#https://www.dcard.tw/_api/forums/pet/posts?popular=false 這是非熱門即最新
#發現他API的網址的邏輯後就可以獲得不同內容了
#可看這篇https://levirve.github.io/blog/2016/Dccard-crawler/

#重要
#如果在改columns的時候直接用:
#latestdata.columns=['標題:','貼文時間:','留言人數:','按讚人數:']會有bug說:
#Pandas doesn't allow columns to be created via a new attribute name
#但是如果建立一個新變數rename2 = ['標題:','貼文時間:','留言人數:','按讚人數:']
#再latestdata.columns=rename2好像就沒事了
#https://stackoverflow.com/questions/52129876/userwarning-pandas-doesnt-allow-columns-to-be-created-via-a-new-attribute-name
#==================================================
import pandas as pd

nopopular = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false')
nopopularjson = json.loads(nopopular.text)

df =pd.DataFrame(nopopularjson,columns =['title' ,'createdAt', 'commentCount','likeCount'])
df
rename = ['標題:','貼文時間:','留言人數:','按讚人數:']
df.columns = rename
df

```

```python
print('熱門文章的「平均留言人數」:',round(popular[('留言人數:')].mean(),2))
print('熱門文章的「平均按讚人數」:',round(popular[('按讚人數:')].mean(),2))
print('非熱門文章的「平均留言人數」:',round(df[('留言人數:')].mean(),2))
print('非熱門文章的「平均按讚人數」:',round(df[('按讚人數:')].mean(),2))
```

```python
import numpy as np
print('熱門文章的「平均留言人數」:',round(np.mean(popular[('留言人數:')]),2))
print('熱門文章的「平均留言人數」:',round(np.mean(popular[('按讚人數:')]),2))
print('熱門文章的「平均留言人數」:',round(np.mean(df[('留言人數:')]),2)) 
print('熱門文章的「平均留言人數」:',round(np.mean(df[('按讚人數:')]),2)) 
```
