## 1. 取出今天所有的新聞

```python
# 先複習一下原本純靜態的爬法

import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

## 2. 取出現在時間兩小時內的新聞

```python
from datetime import datetime,timedelta

two_hours_ago =datetime.now()-timedelta(hours=2)
print(two_hours_ago)

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")
two_hours_data = []
for d in soup.find(class_="part_list_2").find_all('h3'):
    if datetime.strptime(d.find(class_="date").text,'%Y/%m/%d %H:%M') > two_hours_ago:
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
        two_hours_data.append([d.find(class_="date").text, d.find_all('a')[-1].text])
```

```python
import pandas as pd

columns = ['日期','二個小時內的文章']
data=pd.DataFrame(two_hours_data,columns = columns)
data
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
from datetime import datetime,timedelta

three_days =datetime.now()-timedelta(3)
print(three_days)
three=datetime.strptime(three_days.strftime("%Y/%m/%d") + ' 15', '%Y/%m/%d %H')
five=datetime.strptime(three_days.strftime("%Y/%m/%d") + ' 17', '%Y/%m/%d %H')
print('三天前下午三點:',three)
print('三天前下午五點:',five)
print(type(three),type(five))
```
